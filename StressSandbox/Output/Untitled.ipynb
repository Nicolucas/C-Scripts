```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.gridspec import GridSpec
from matplotlib import rc
```

```python
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
rc('text', usetex=True)
```

```python
def format_axes(fig):
    for i, ax in enumerate(fig.axes):
        ax.tick_params(labelbottom=True, labelleft=True)
        
def PlotRow(FileName, XLabel, YLabel, title, axis):
    FrictionData = pd.read_csv(FileName,sep=";",header=None,names=["time", "Traction", "Slip","Theta", "Fric","SlipDot"])
    FrictionData.plot(x=XLabel[0], y=YLabel[0], title = title, ax=axis)
    axis.set_xlabel(XLabel[1])
    axis.set_ylabel(YLabel[1])
    
def PlotRows(FileName,ListOfAxes,title):
    PlotRow(FileName, ["time","$time$"], ["Traction","$Traction$"], title, ListOfAxes[0])
    PlotRow(FileName, ["time","$time$"], ["Slip","$Slip$"], title, ListOfAxes[1])
    PlotRow(FileName, ["Slip","$Slip$"], ["SlipDot","$SlipDot$"], title, ListOfAxes[2])
    PlotRow(FileName, ["Slip","$Slip$"], ["Fric","$Friction$"], title, ListOfAxes[3])


```

```python
ListFilenames=["../Output/Drama.txt"]
Titles = [""]




fig = plt.figure(constrained_layout=True, figsize=(9,5))

gs = GridSpec(1, 4, figure=fig)
for j in range(1):
    ax1 = [fig.add_subplot(gs[j, i]) for i in range(4)]
    PlotRows(ListFilenames[j],ax1,Titles[j])



fig.suptitle("Friction Laws: Rate and State")
format_axes(fig)

plt.show()
```
