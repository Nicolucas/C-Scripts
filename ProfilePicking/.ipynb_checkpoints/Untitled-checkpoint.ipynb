```python
import os
import numpy as np
from matplotlib import pyplot as plt
from Lib_ProfileFromImage import *
from se2waveload import *
from glob import glob
import pandas as pd

plt.style.use('seaborn-whitegrid')
from matplotlib import cm

cmap = cm.get_cmap("plasma_r")


import time
start_time = time.time()
```

```python
def PlotTimeProfileSet(ax, Set,SlipSlipRate,title):
    UnitsDict = {"Slip" : "Slip [m]", "SlipRate" : "Slip Rate [m/s]"}
    
    
    ax.set(xlabel = 'Time [s]', ylabel = UnitsDict[SlipSlipRate], 
           title = title)
    
    if (SlipSlipRate == "Slip"):
        for item in Set:
            ax.plot(item.Time, item.DispX, c = cmap(item.Coord[0]/8000), label = "{x}m".format(x = item.Coord[0]))
    elif (SlipSlipRate == "SlipRate"):
        for item in Set:
            ax.plot(item.Time, item.VelX, c = cmap(item.Coord[0]/8000), label = "{x}m".format(x = item.Coord[0]))
    
    ax.legend()
    
def PlotReference(ax,SlipSlipRate):
    FileNameDict = {"Slip" : "slip", "SlipRate" : "sr"}
    ListReference = sorted(glob(path+"Reference/digitizerdata/crack-{}-*.txt".format(FileNameDict[SlipSlipRate])))

    for idx, item in enumerate(ListReference):
        line = pd.read_csv(item,header=None)
        ax.plot(line[0],line[1],c = cmap(idx/5), label = "", ls = "--")
        
def PlotHandPickedReference(ax,SlipSlipRate):
    ListReference = sorted(glob("Output/HandPicking/{}/".format(SlipSlipRate)+"*.csv"))

    for idx, item in enumerate(ListReference):
        line = pd.read_csv(item)
        ax.plot(line["X"],line["Y"],c = cmap(idx/5), label = "", ls = "--")
        
def LoadAndPlot(SingleFile,SlipSlipRate):
    head, tail = os.path.split(SingleFile)
    File = LoadPickleFile(Filename = tail,FolderPath = head+"/")

    try:
      fig = plt.figure(figsize = (10,5), constrained_layout=True)
      gs = fig.add_gridspec(1, 1)
      ax = fig.add_subplot(gs[:, :])
    except:
      fig = plt.figure(figsize = (10,5))
      ax = fig.add_subplot(1,1,1)

    ax.set_xlim([0,4])
    ax.set_ylim([-1,10])

    PlotHandPickedReference(ax, SlipSlipRate)
    PlotTimeProfileSet(ax, File, SlipSlipRate, head[-20:])
    
def AllInOnePlot(ListOfFiles,SlipSlipRate):
    try:
      fig = plt.figure(figsize = (10,5), constrained_layout=True)
      gs = fig.add_gridspec(1, 1)
      ax = fig.add_subplot(gs[:, :])
    except:
      fig = plt.figure(figsize = (10,5))
      ax = fig.add_subplot(1,1,1)

    ax.set_xlim([0,4])
    ax.set_ylim([-1,10])
    
    for SingleFile in ListOfFiles:
        head, tail = os.path.split(SingleFile)
        File = LoadPickleFile(Filename = tail,FolderPath = head+"/")


        PlotReference(ax, SlipSlipRate)
        PlotTimeProfileSet(ax, File, SlipSlipRate, head[-20:])
```

```python
def CalculateKostrovSolution(Time,Stress,mu, v,r):
    return (Stress/mu)*np.sqrt((v**2)*(Time**2)-r**2)
```

```python
TimeList = np.arange(0, 4, 0.01).tolist()
```

```python
Slip = [CalculateKostrovSolution(Time,20,0.5, 2000,4000) for Time in TimeList]
plt.plot(TimeList,Slip)
```
