```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.gridspec import GridSpec
from matplotlib import rc
```

```python
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
rc('text', usetex=True)
```

```python
def format_axes(fig):
    for i, ax in enumerate(fig.axes):
        ax.tick_params(labelbottom=True, labelleft=True)
        
def PlotRow(FileName, XLabel, YLabel, title, axis):
    FrictionData = pd.read_csv(FileName,sep=";",header=None,names=["time", "Slip", "SlipDot","Theta", "Friction","Traction"])
    FrictionData.plot(x=XLabel[0], y=YLabel[0], title = title, ax=axis)
    axis.set_xlabel(XLabel[1])
    axis.set_ylabel(YLabel[1])
    
def PlotRows(FileName,ListOfAxes,title):
    PlotRow(FileName, ["time","$time$"], ["Slip","$Slip$"], title, ListOfAxes[0])
    PlotRow(FileName, ["time","$time$"], ["Traction","$Traction$"], title, ListOfAxes[1])
    PlotRow(FileName, ["time","$time$"], ["Friction","$Friction$"], title, ListOfAxes[2])
    PlotRow(FileName, ["Slip","$Slip$"], ["Friction","$Friction$"], title, ListOfAxes[3])

```

```python
ListFilenames=["../Output/Demo_01.txt", "../Output/Demo_02.txt", "../Output/Demo_03.txt"]
Titles = ["","",""]


fig = plt.figure(constrained_layout=True, figsize=(9,5))

gs = GridSpec(4, 4, figure=fig)
for j in range(3):
    ax1 = [fig.add_subplot(gs[j, i]) for i in range(4)]
    PlotRows(ListFilenames[j],ax1,Titles[j])



fig.suptitle("Friction Laws: Linear Slip Weakening")
format_axes(fig)

plt.savefig("../Output/Linear.png")

plt.show()
```
