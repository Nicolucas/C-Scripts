```python
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.gridspec import GridSpec


from matplotlib import rc
```

```python
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
rc('text', usetex=True)
```

```python
def format_axes(fig):
    for i, ax in enumerate(fig.axes):
        ax.tick_params(labelbottom=True, labelleft=True)
        
def PlotRow(FileName, XLabel, YLabel, title, axis):
    FrictionData = pd.read_csv(FileName,sep=";",header=None,names=["time", "Tau", "Slip", "SlipRate", "Fric","Theta"])
    FrictionData.plot(x=XLabel[0], y=YLabel[0], title = title, ax=axis)
    axis.set_xlabel(XLabel[1])
    axis.set_ylabel(YLabel[1])
    
def PlotRows(FileName,ListOfAxes,title):
    PlotRow(FileName, ["time","$Time$"], ["Slip","$Slip$"], title, ListOfAxes[0])
    PlotRow(FileName, ["time","$Time$"], ["SlipRate","$Slip$ $Rate$"], title, ListOfAxes[1])
    PlotRow(FileName, ["time","$Time$"], ["Fric","$\mu$"], title, ListOfAxes[2])
    PlotRow(FileName, ["Slip","$Slip$"], ["Fric","$\mu$"], title, ListOfAxes[3])
    PlotRow(FileName, ["Slip","$Slip$"], ["Theta","$\Theta$"], title, ListOfAxes[4])
```

```python
ListFilenames=["../Output/TestLSW.txt","../Output/TestRS_HealingLaw_NoThetaUp.txt","../Output/TestModRS_HealingLaw.txt",
               "../Output/TestRS_HealingLaw.txt","../Output/TestRS_SlipLaw.txt","../Output/TestRS_PRZLaw.txt"]
Titles = ["Linear Slip Weakening", "Dieterich-Ruina Rate and State", "Mod Rate and State",
         "Dieterich Law","Ruina Law", "PRZ Law"]




fig = plt.figure(constrained_layout=True, figsize=(20,10))

gs = GridSpec(6, 5, figure=fig)
for j in range(6):
    ax1 = [fig.add_subplot(gs[j, i]) for i in range(5)]
    PlotRows(ListFilenames[j],ax1,Titles[j])



fig.suptitle("Friction Laws")
format_axes(fig)

fig.savefig("../Output/Saw.png")

plt.show()
```
